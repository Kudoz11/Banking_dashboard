```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df= pd.read_csv('Banking.csv')
df.head()
```

```python
#check the shape of data frame
df.describe()
```

```python
bins=[0,100000,300000,float('inf')]
labels=['Low','Medium','High']
df['Income Band']=pd.cut(df['Estimated Income'],bins=bins,labels=labels,right=False)
```

```python
#Income Band= SWITCH(TRUE),'Clie'
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
categorical_cols=df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns

for col in categorical_cols:
  print(f"value counts for '{col}':")
  display(df[col].value_counts())# .plot(kind='bar')
  print("\n")
```

```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor)
```

```python
for i, predictor in enumerate(df[['BRId','GenderId','IAId','Amount of Credit Cards','Nationality','Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting','Income Band']].columns):
  plt.figure(i)
  sns.countplot(data=df,x=predictor,hue='Nationality')
```

```python
# histoplot  of value counts foor  different occupation
for col in categorical_cols:
  if col=="Occupation":
    continue
  plt.figure(figsize=(8,4))
  sns.histplot(df[col])
  plt.title('histogram of of Occupation Count')
  plt.xlabel(col)
  plt.ylabel("Count")
  plt.show()
  print("/n")
```

```python
# numerical analytics:
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']

# Univariate analyis and visulation
plt.figure(figsize=(15,10))
for i ,col in enumerate(numerical_cols):
  plt.subplot(4,3,i+1)
  sns.histplot(df[col],kde=True)
  plt.title(col)
  plt.show()
  print("/n")
```

```python
numerical_cols=['Estimated Income','Superannuation Savings','Credit Card Balance','Bank Loans','Bank Deposits','Checking Accounts','Saving Accounts','Foreign Currency Account','Business Lending']
correlation_matrix=df[numerical_cols].corr()
plt.figure(figsize=(12,12))
sns.heatmap(correlation_matrix,annot=True,cmap='crest',fmt=".2f")
plt.title('Correlation Matrix')
plt.show()
```

# insights of EDA:
 1. The strongest positive correlation occur among "Bank Deposits" with "Checking Accounts","Saving Accounts" and "Foreign Currency Account" indicating that customers who maintain balance in one account type often substantial amount/funds acros other hold substantial amountt/fund
